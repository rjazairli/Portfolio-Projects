```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe_tracks = pd.read_csv('C:/Users/ramie/Downloads/Spotify Datasets/tracks.csv')
dataframe_tracks.head()
```

```python
# Checking null values

pd.isnull(dataframe_tracks).sum()
```

```python
# Check for total number of rows and columns in dataset, check datatypes, and memory usage
dataframe_tracks.info()
```

```python
# least popular songs
sorted_df = dataframe_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
# View descriptive statistics for the numerical variables that are present in our columns
dataframe_tracks.describe().transpose()
```

```python
# Check for the Top 10 songs:
most_popular = dataframe_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
# Set the index of the column to be "Release Date"
dataframe_tracks.set_index("release_date", inplace = True)
dataframe_tracks.index = pd.to_datetime(dataframe_tracks.index)
dataframe_tracks.head()
```

```python
# Check the artist in a specific row of the dataset using index location 
dataframe_tracks[["artists"]].iloc[18]
```

```python
# Converting song duration from ms to s:
dataframe_tracks["duration"]= dataframe_tracks["duration_ms"].apply(lambda x: round(x/1000))
dataframe_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
dataframe_tracks.duration.head()
```

```python
# Creating visualization (Heat Map) using correlations:
corr_df = dataframe_tracks.drop(["key", "mode", "explicit"], axis = 1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# Observations of HeatMap above:
# 1. Loudness has a strong positive correlation to energy 
# 2. Acousticness has a strong negative correlation to energy
# 3. Danceability has a moderately strong positive corelation to valence
# 4(a). The duration of a song typically has little to no correlation with other variables such as popularity
# 4(b). For example, the song "Bohemian Rhapsody - Queen" is almost 6 minutes long and was touted by 
# radio hosts for being unplayable in radio stations due to the song's length. However, the song topped
# the UK Singles Chart for 9 weeks after its release, with over 150 million record sales.
```

```python
sample_df = dataframe_tracks.sample(int(0.004*len(dataframe_tracks)))
```

```python
print(len(sample_df))
```

```python
# Creating a Linear Regression plot between Loudness and Energy
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

```python
# Creating a Linear Regression plot between Popularity and Acousticness
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "b").set(title = "Popularity vs Acousticness Correlation")
```

```python
# Observations:
# 1. If the acousticness of a song increases, the popularity of the song decreases
# 2. Vice versa
```

```python
dataframe_tracks['dates'] = dataframe_tracks.index.get_level_values('release_date')
dataframe_tracks.dates = pd.to_datetime(dataframe_tracks.dates)
years = dataframe_tracks.dates.dt.year
```

```python
# Displaying Bar Chart of songs released by year
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
#Creating a Bar Chart illustrating song duration by year
total_dr = dataframe_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation=60)
```

```python
# Observations:
# 1. It appears that song duration steadily increases or decreases from the early 1900s up until 2010
# 2. After 2010, it appears that the duration of songs have gradually decreased
```
